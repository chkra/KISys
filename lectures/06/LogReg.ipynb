# Logistische Regression als Klassifikator mit Merkmalsauswahl

Der folgende Code zeigt, wie die Gewichte (Koeffizienten) eines logistischen Regressionsmodells ermittelt und visualisiert werden können. Diese Gewichte geben an, wie stark und in welche Richtung (positiv oder negativ) die einzelnen Merkmale des Datensatzes die Zielvariable beeinflussen.

Wir verwenden den **Iris-Datensatz**, einen bekannten Datensatz in der maschinellen Lernforschung, der Informationen über Blumenarten enthält. Die Zielvariable wird für eine binäre Klassifikation angepasst: Wir unterscheiden die Klasse "Setosa" (Wert 1) von den anderen Arten (Wert 0).

Der Code besteht aus den folgenden Schritten:

1. **Datensatz laden:** Die Daten werden mit sklearn.datasets geladen und in Merkmale (X) und Zielvariable (y) aufgeteilt.
2. **Modell trainieren:** Eine logistische Regression wird auf den Daten trainiert, um die Gewichte zu bestimmen.
Gewichte ausgeben: Die Gewichte der einzelnen Features werden in Textform ausgegeben und geben an, wie wichtig jedes Feature ist.
3. **Visualisierung:** Ein Balkendiagramm zeigt die Gewichte der Features in aufsteigender Reihenfolge. Dies bietet eine intuitive Darstellung der Merkmalswichtigkeit.

All Credit to ChatGPT :-)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

# 1. Iris-Datensatz laden
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = (iris.target == 0).astype(int)  # Binäre Klassifikation: Setosa (1) vs. andere (0)

# 2. Logistische Regression trainieren
logreg = LogisticRegression(max_iter=200)
logreg.fit(X, y)

# 3. Gewichtsvektor ausgeben
weights = pd.Series(logreg.coef_[0], index=X.columns)
print("Gewichtsvektor der logistischen Regression:")
print(weights.sort_values(ascending=False))

# 4. Gewichte als Balkendiagramm plotten
weights.sort_values(ascending=True).plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Gewichtsvektor der logistischen Regression")
plt.xlabel("Gewicht")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
